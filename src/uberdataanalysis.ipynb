DATA PREPROCESSING

```python
df['PURPOSE']=df['PURPOSE'].fillna("Unknown") # Fill missing values in 'PURPOSE' column with "Unknown"
df.head()
```

```python
#converting the 'START_DATE' and 'END_DATE' column to datetime format
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors='coerce')
df['END_DATE'] = pd.to_datetime(df['END_DATE'], errors='coerce')
df.info()
```

```python
from datetime import datetime
df['Date']=pd.DatetimeIndex(df['START_DATE']).date 
df['Time']=pd.DatetimeIndex(df['START_DATE']).hour
df.head(5)
```

```python
df['day_night']=pd.cut(df['Time'],bins=[0,10,15,19,24],labels=['morning','afternoon','evening','night'])
df.head(5)
```
